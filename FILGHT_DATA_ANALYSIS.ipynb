```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as smt
df = pd.read_csv(r"C:\Users\Lenovo\Desktop\Flight_Data (1).csv")
p = df.groupby('departure_time')['price'].mean()
df1 = df[df['price']<800000]
p1 = df1.groupby('departure_time')['price'].mean()
p1.plot(kind = "bar")
plt.xlabel('TIME OF DEPARTURE')
plt.ylabel('PRICE')
plt.title('AVERAGE PRICE W.R.T TIME OF DEPARTURE')
plt.grid(axis = 'y')
plt.show()
```
